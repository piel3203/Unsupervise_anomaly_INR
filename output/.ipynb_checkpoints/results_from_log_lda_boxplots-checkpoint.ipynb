```python
import pandas as pd
import os
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from collections import Counter
from ast import literal_eval
from statsmodels.multivariate.manova import MANOVA
from IPython.display import display
from sklearn.decomposition import PCA
from scipy.stats import shapiro
import statsmodels.api as sm
from statsmodels.formula.api import mixedlm
#https://www.statsmodels.org/stable/generated/statsmodels.multivariate.manova.MANOVA.html
```

```python
# ----------------------------
# CONFIG
# ----------------------------
muscle = 'RF'
n_folds = 1
output_csv = f"fisher_scores_and_metrics_{muscle}.csv"
output_dir = f"plots_{muscle}"
os.makedirs(output_dir, exist_ok=True)
```

```python
# ----------------------------
# UTILS
# ----------------------------
def load_subjects(file_path, group=None):
    with open(file_path, 'r') as f:
        subjects = [line.strip().lower() for line in f]
    if group:
        return pd.DataFrame({'Subject': subjects, 'Group': group})
    return subjects

def categorize_subjects(subjects):
    categorized = []
    for subj in subjects:
        if subj.startswith('sujet'):
            categorized.append({'Subject': subj, 'Group': 'healthy_subjects'})
        else:
            categorized.append({'Subject': subj, 'Group': 'patients_wo_sarcopenia'})
    return pd.DataFrame(categorized)

def parse_log_file(path):
    parsed_data = []
    if not os.path.isfile(path):
        return pd.DataFrame()
    with open(path, 'r') as file:
        lines = file.readlines()
    batch_cases, z_latents_batch_str, stats = [], "", {}
    stats_pattern = r"ASD: ([\d.]+), HSD: ([\d.]+), HSD95: ([\d.]+), DSC: ([\d.]+), err_vol_cm_3: ([\d.]+), err_vol_percent: ([\d.]+)"
    i = 0
    while i < len(lines):
        line = lines[i]
        if "Batch cases: " in line:
            batch_cases = line.strip().replace("Batch cases: ", "").split(', ')
        elif batch_cases and "z:" in line:
            tensor_lines = [line.strip().split('z:')[1].strip()]
            i += 1
            while i < len(lines) and ("tensor" in lines[i] or "  " in lines[i]):
                tensor_lines.append(lines[i].strip())
                i += 1
            z_latents_batch_str = ' '.join(tensor_lines)
            continue
        elif batch_cases and "Batch ASD:" in line:
            match = re.search(stats_pattern, line)
            stats = {k: float(v) for k, v in zip(
                ['ASD','HSD','HSD95','DSC','Err_Vol_CM3','Err_Vol_Percent'], match.groups()
            )} if match else {k: None for k in ['ASD','HSD','HSD95','DSC','Err_Vol_CM3','Err_Vol_Percent']}
        if batch_cases and stats:
            for case in batch_cases:
                parsed_data.append((case.strip().lower(), z_latents_batch_str, *stats.values()))
            batch_cases, z_latents_batch_str, stats = [], "", {}
        i += 1
    columns = ['Subject','z_latents_batch','ASD','HSD','HSD95','DSC','Err_Vol_CM3','Err_Vol_Percent']
    return pd.DataFrame(parsed_data, columns=columns)

def convert_tensor_string_to_array(tensor_str):
    tensor_str = re.sub(r'Parameter containing:', '', tensor_str).strip()
    tensor_str = tensor_str.replace("tensor(", "").split(", device=")[0]
    try:
        return np.array(literal_eval(tensor_str))
    except:
        return None
```

```python

# ----------------------------
# MAIN LOOP FOLDS
# ----------------------------
results = []
df_combined_list = []

colors = {
    'patient_with_sarcopenia': 'red',
    'patients_wo_sarcopenia': 'green',
    'healthy_subjects': 'blue'
}
markers = {'Train': 'x', 'Test': 'o'}
group_name_mapping = {
    'healthy_subjects': 'young_adults',
    'patients_wo_sarcopenia': 'older_adults_wo_sarcopenia',
    'patient_with_sarcopenia': 'older_adults_with_sarcopenia',
}

fold_palette = sns.color_palette("Set1", n_colors=n_folds)

for fold in range(1, n_folds + 1):
    print(f"Processing fold {fold}...")

    path_sarcopenia = f'../casename_files/DIASEM/{muscle}/sarcopenia_subj_DIASEM.txt'
    path_train = f'../casename_files/DIASEM/{muscle}/train_cases_{fold}.txt'
    path_test = f'../casename_files/DIASEM/{muscle}/test_cases_{fold}.txt'

    log_paths = {
        'test': f'./DIASEM_{muscle}_set{fold}_NO_MISMATCH_registered_turned_1500_eval_ax2_x1_healthy/log.txt',
        'sarcopenia': f'./DIASEM_{muscle}_set{fold}_NO_MISMATCH_registered_turned_1500_eval_ax2_x1_sarcopenia/log.txt',
        'train': f'./DIASEM_{muscle}_set{fold}_NO_MISMATCH_registered_turned_1500_eval_ax2_x1_train/log.txt'
    }

    sarcopenia_df = load_subjects(path_sarcopenia, 'patient_with_sarcopenia')
    train_df = categorize_subjects(load_subjects(path_train))
    test_df = categorize_subjects(load_subjects(path_test))

    log_test = parse_log_file(log_paths['test'])
    log_sarc = parse_log_file(log_paths['sarcopenia'])
    log_train = parse_log_file(log_paths['train'])

    for df in [log_test, log_sarc, log_train]:
        if not df.empty:
            df['Subject'] = df['Subject'].str.strip("[]").str.replace("'", "")
            df['z_latents_batch'] = df['z_latents_batch'].apply(convert_tensor_string_to_array)

    log_test = log_test.merge(test_df, on='Subject', how='left').assign(Set='Test')
    log_sarc = log_sarc.merge(sarcopenia_df, on='Subject', how='left').assign(Set='Test')
    log_train = log_train.merge(train_df, on='Subject', how='left').assign(Set='Train')

    # Concatène tous les sujets train et test
    combined_df = pd.concat([log_test, log_sarc, log_train], ignore_index=True)

    shapes = combined_df['z_latents_batch'].apply(lambda x: x.shape if x is not None else None)
    if shapes.dropna().empty:
        continue

    most_common_shape = Counter(shapes.dropna()).most_common(1)[0][0]
    df_filtered = combined_df[combined_df['z_latents_batch'].apply(lambda x: x is not None and x.shape == most_common_shape)]
    if df_filtered.empty:
        continue

    X = np.stack(df_filtered['z_latents_batch'].values)
    if X.ndim == 3:
        X = X.reshape(X.shape[0], -1)
    y = df_filtered['Group'].values  # Labels (Y, OS et OH)

    # ----------------------------
    # 1) PCA + MANOVA sur le latent space
    # ----------------------------
    n_components = min(64, X.shape[1])  # max 32 composantes
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X)

    # Crée DataFrame PCA
    pca_cols = [f"PC{i+1}" for i in range(X_pca.shape[1])]
    df_latents = pd.DataFrame(X_pca, columns=pca_cols)
    df_latents["Group"] = y

    # MANOVA
    formula = " + ".join(pca_cols) + " ~ Group"
    maov = MANOVA.from_formula(formula, data=df_latents)
    manova_res = maov.mv_test()
    p_value = manova_res.results['Group']['stat']['Pr > F']['Wilks\' lambda']
    print(f"Fold {fold} MANOVA p-value:", p_value)

    # ----------------------------
    # 2) LDA pour la visualisation + Fisher score
    # ----------------------------
    lda = LDA(n_components=2)
    z_lda = lda.fit_transform(X, y)

    # Calcul Fisher Score (sur les coordonnées LDA)
    classes = np.unique(y)
    overall_mean = np.mean(z_lda, axis=0)
    S_B, S_W = np.zeros((2,2)), np.zeros((2,2))
    for cls in classes:
        X_c = z_lda[y == cls]
        mean_c = np.mean(X_c, axis=0)
        n_c = X_c.shape[0]
        mean_diff = (mean_c - overall_mean).reshape(-1, 1)
        S_B += n_c * (mean_diff @ mean_diff.T)
        S_W += np.cov(X_c, rowvar=False) * (n_c - 1)
    fisher_score_multi = np.trace(S_B) / np.trace(S_W)

    # Sauvegarde des résultats (Fisher + p-value MANOVA)
    results.append({'fold': fold, 'fisher_score': fisher_score_multi, 'p_value': p_value})

    if p_value < 0.001:
        titre_p_value = "p-value < 0.001"
    else:
        titre_p_value = f"p-value = {p_value:.4f}"
    # ----------------------------
    # Plot LDA
    # ----------------------------
    plt.figure(figsize=(10, 9))
    labels_added = set()
    for idx, row in df_filtered.iterrows():
        marker = markers.get(row['Set'], 'o')
        color = colors.get(row['Group'], 'gray')
        group_label = group_name_mapping.get(row['Group'], row['Group'])
        if group_label not in labels_added:
            plt.scatter(z_lda[idx,0], z_lda[idx,1], c=color, marker=marker, alpha=0.7, label=group_label)
            labels_added.add(group_label)
        else:
            plt.scatter(z_lda[idx,0], z_lda[idx,1], c=color, marker=marker, alpha=0.7)
    for set_type, marker in markers.items():
        plt.scatter([], [], color='black', marker=marker, label=set_type)
    plt.xlabel('LDA Component 1', fontsize=16)
    plt.ylabel('LDA Component 2', fontsize=16)
    plt.legend(title="Group and Set", fontsize=12)
    plt.title(f'LDA fold {fold}\nFisher score: {fisher_score_multi:.2f}\np-value (MANOVA): {titre_p_value}', fontsize=16)
    plt.grid(True)
    plt.show()
    plt.savefig(os.path.join(output_dir, f"lda_fold_{fold}.png"))

    # ----------------------------
    # Metrics pour boxplots
    # ----------------------------
    for test_set, log_file, label in zip(['healthy','sarcopenia'], [log_paths['test'], log_paths['sarcopenia']], ['healthy','sarcopenia']):
        df_metrics = parse_log_file(log_file)
        if not df_metrics.empty:
            df_metrics['fold'] = fold
            df_metrics['source'] = f'Fold_{fold}_Test_Set_{label}'
            df_combined_list.append(df_metrics)

# ----------------------------
# Combine metrics and results
# ----------------------------
df_combined = pd.concat(df_combined_list, ignore_index=True)
df_results = pd.DataFrame(results)  # contient Fisher + p-values
df_final = df_combined.merge(df_results, left_on='fold', right_on='fold', how='left')
df_final.to_csv(output_csv, index=False)
print("Résultats sauvegardés dans", output_csv)
display(df_final)
```

```python
# ==========================================
# ROC analysis on DSC between healthy / sarcopenia
# ==========================================

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# 1️⃣ Extraire les données nécessaires
# (On réutilise df_all, qui contient toutes les valeurs DSC et la colonne 'Group')
df_roc = df_all.copy()

# S'assurer que seules les deux classes d'intérêt sont présentes
df_roc = df_roc[df_roc["Group"].isin(["healthy", "sarcopenia"])]

# Convertir les groupes en valeurs binaires : 0 = healthy, 1 = sarcopenia
y_true = (df_roc["Group"] == "sarcopenia").astype(int).values
y_score = df_roc["DSC"].values  # plus DSC est élevé, plus c’est "healthy"

# ⚠️ Comme le DSC est inversement lié à la sévérité, on inverse le score
# (car un DSC plus faible correspond à un sujet sarcopénique)
y_score_inverted = 1 - y_score

# 2️⃣ Calcul de la courbe ROC
fpr, tpr, thresholds = roc_curve(y_true, y_score_inverted)
roc_auc = auc(fpr, tpr)

# 3️⃣ Trouver le seuil optimal (indice de Youden)
youden_index = np.argmax(tpr - fpr)
optimal_threshold = thresholds[youden_index]
optimal_sensitivity = tpr[youden_index]
optimal_specificity = 1 - fpr[youden_index]

print(f"AUC = {roc_auc:.4f}")
print(f"Optimal threshold (on inverted DSC) = {optimal_threshold:.4f}")
print(f"Sensitivity = {optimal_sensitivity:.3f}, Specificity = {optimal_specificity:.3f}")

# 4️⃣ Tracé de la courbe ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.scatter(fpr[youden_index], tpr[youden_index], color='red', s=80, label=f'Optimal threshold = {optimal_threshold:.3f}')
plt.text(fpr[youden_index]+0.03, tpr[youden_index]-0.05, f'Th={optimal_threshold:.3f}', fontsize=10)

plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12)
plt.title('ROC Curve for DSC Classification\n(Healthy vs Sarcopenia)', fontsize=14)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)

filename = os.path.join(output_dir, f"ROC_{metric}.pdf")
plt.savefig(filename)
plt.show()

# 5️⃣ Interprétation :
# Le seuil trouvé correspond au DSC qui sépare au mieux les deux groupes :
#  - si DSC > (1 - threshold) → prédire healthy
#  - sinon → prédire sarcopenia
```

```python
# ==========================================
# Ajouter la ligne de seuil ROC sur les boxplots DSC
# ==========================================

# Seuil calculé précédemment avec la ROC
dsc_threshold = 1 - optimal_threshold  # rappel : on avait inversé le DSC

print(f"Seuil global DSC (healthy vs sarcopenia) = {dsc_threshold:.4f}")

# On redessine uniquement le boxplot pour le DSC avec la ligne de seuil
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(
    x='source', y='DSC', data=df_combined,
    palette=fold_colors, width=0.5, saturation=0.5
)

# Ajouter les points individuels
for i, source in enumerate(df_combined['source'].unique()):
    source_data = df_combined[df_combined['source'] == source]
    sns.stripplot(
        x='source', y='DSC', data=source_data,
        color=fold_palette[i//2], jitter=True, size=5, alpha=0.9
    )

# Ligne de seuil horizontal
plt.axhline(y=dsc_threshold, color='red', linestyle='--', lw=2,
            label=f'Seuil ROC (DSC = {dsc_threshold:.3f})')

# Labels et style
plt.xticks(
    ticks=range(len(new_xtick_labels)),
    labels=new_xtick_labels, rotation=45, ha='right'
)
plt.ylabel("DSC", fontsize=12)
plt.title(f"Boxplot DSC across folds\n(Threshold = {dsc_threshold:.3f}, global ROC)", fontsize=14)
plt.legend(loc='lower right')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Sauvegarde PDF
plt.savefig(os.path.join(output_dir, f"boxplot_DSC_with_threshold.pdf"))
plt.show()
```

```python
# ==========================================
# ROC GLOBALE (tous folds combinés)
# ==========================================

from sklearn.metrics import roc_curve, auc, confusion_matrix, accuracy_score, recall_score, precision_score

# On réutilise df_all (déjà concaténé)
df_global = df_all[df_all["Group"].isin(["healthy", "sarcopenia"])].copy()

# Encodage binaire des labels
y_true = (df_global["Group"] == "sarcopenia").astype(int).values
y_score = df_global["DSC"].values

# ⚠️ On inverse le DSC (faible DSC = sarcopenia)
y_score_inv = 1 - y_score

# Calcul ROC
fpr, tpr, thresholds = roc_curve(y_true, y_score_inv)
roc_auc = auc(fpr, tpr)

# Seuil optimal (indice de Youden)
youden_index = np.argmax(tpr - fpr)
optimal_threshold_global = thresholds[youden_index]
dsc_threshold_global = 1 - optimal_threshold_global  # on revient à l’échelle du DSC

print(f"✅ global ROC : AUC = {roc_auc:.4f}")
print(f"✅ Optimal Threshold DSC = {dsc_threshold_global:.4f}")
print(f"   Sensitivity (TPR) = {tpr[youden_index]:.3f}")
print(f"   Spécificity (1-FPR) = {1 - fpr[youden_index]:.3f}")

# Tracé de la courbe ROC globale
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.scatter(fpr[youden_index], tpr[youden_index], color='red', s=80,
            label=f'Optimal Threshold DSC = {dsc_threshold_global:.3f}')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.title('Global ROC Curve (All folds combined)')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ==========================================
# Évaluation par fold avec le seuil global
# ==========================================

fold_metrics = []

for fold in range(1, 6):
    # Récupérer les deux sous-groupes
    dsc_healthy = dfs[f"Fold_{fold}_healthy"]
    dsc_sarcopenia = dfs[f"Fold_{fold}_sarcopenia"]

    df_fold = pd.concat([
        dsc_healthy.assign(Group="healthy"),
        dsc_sarcopenia.assign(Group="sarcopenia")
    ])

    # Vecteurs
    y_true = (df_fold["Group"] == "sarcopenia").astype(int).values
    y_pred = (df_fold["DSC"] < dsc_threshold_global).astype(int)  # prédiction basée sur le seuil

    # Calcul des métriques
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    acc = accuracy_score(y_true, y_pred)
    sens = recall_score(y_true, y_pred)  # sensibilité = TPR
    spec = tn / (tn + fp) if (tn + fp) > 0 else np.nan
    prec = precision_score(y_true, y_pred)
    auc_fold = auc(*roc_curve(y_true, 1 - df_fold["DSC"].values)[:2])

    fold_metrics.append({
        "Fold": fold,
        "Accuracy": acc,
        "Sensitivity": sens,
        "Specificity": spec,
        "Precision": prec,
        "AUC": auc_fold
    })

# Résumé global
df_fold_metrics = pd.DataFrame(fold_metrics)
print("✅ Résumé des performances par fold :")
display(df_fold_metrics)
print("Moyenne ± écart-type :")
display(df_fold_metrics.describe().loc[["mean", "std"]])
```
