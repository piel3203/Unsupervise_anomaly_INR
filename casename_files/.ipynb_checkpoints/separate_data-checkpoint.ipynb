```python
for i in range(2,19):
    # Path to your folder
    data_folder = '../data/labels_separated/label{}/'.format(i)
    output_folder='./FULGUR/label{}/'.format(i)
    # Create the folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Get all filenames in the folder that match the format "FULGUR_***_******_label1.nii.gz"
    filenames = [f for f in os.listdir(data_folder) ]
    print (filenames)

    # Extract the IDs by removing the ".nii.gz" extension
    ids = [f.replace('.nii.gz', '') for f in filenames]

    # Split into train and test (10% test)
    train_ids, test_ids = train_test_split(ids, test_size=0.1, random_state=42)

    # Write the train set to a text file
    with open(os.path.join(output_folder, 'train_cases.txt'), 'w') as train_file:
        for train_id in train_ids:
            train_file.write(train_id + '\n')

    # Write the test set to a text file
    with open(os.path.join(output_folder, 'test_cases.txt'), 'w') as test_file:
        for test_id in test_ids:
            test_file.write(test_id + '\n')

    print("Train and test sets have been written to 'train_cases.txt' and 'test_cases.txt'.")
```

```python
# Path to your folder
data_folder = '../data/data_DIASEM/VL/labels/'
output_folder='./DIASEM/VL/'
# Create the folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

    
```

```python
# Get all filenames in the folder that match the format "FULGUR_***_******_label1.nii.gz"
filenames = [f for f in os.listdir(data_folder) if f.endswith('.nii') or f.endswith('.nii.gz')]
#print (filenames)

# Extract the IDs by removing the ".nii.gz" extension
ids = [f.replace('.nii.gz', '') for f in filenames]
print(ids)
# Split into train and test (10% test)
train_ids, test_ids = train_test_split(ids, test_size=0.1, random_state=42)
```

```python
# Write the train set to a text file
with open(os.path.join(output_folder, 'train_cases.txt'), 'w') as train_file:
    for train_id in train_ids:
        train_file.write(train_id + '\n')

# Write the test set to a text file
with open(os.path.join(output_folder, 'test_cases.txt'), 'w') as test_file:
    for test_id in test_ids:
        test_file.write(test_id + '\n')

print("Train and test sets have been written to 'train_cases.txt' and 'test_cases.txt'.")
```
