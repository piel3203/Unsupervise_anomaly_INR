```python
# Spécifie le chemin de ton fichier texte
input_file = "RF_all_suj.txt"
output_file = "RF_all_suj_ordonnes.txt"

# Lire les lignes du fichier
with open(input_file, "r") as f:
    lignes = f.readlines()

# Supprimer les lignes vides et les espaces inutiles, puis trier
lignes = [ligne.strip() for ligne in lignes if ligne.strip()]
lignes.sort()  # Tri alphabétique

# Écrire le résultat dans un nouveau fichier
with open(output_file, "w") as f:
    for ligne in lignes:
        f.write(ligne + "\n")

print(f"Les sujets ont été triés et enregistrés dans '{output_file}'.")
```
