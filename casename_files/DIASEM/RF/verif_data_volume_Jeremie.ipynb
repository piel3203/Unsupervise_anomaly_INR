```python
import pandas as pd

# Load the CSV and Excel
final_folds_df = pd.read_csv("final_combined_df.csv")
modif_df = pd.read_excel("MODIF_table_donnees_DIASEM.xlsx")

# Ensure ID columns are strings
final_folds_df["ID"] = final_folds_df["ID"].astype(str)
modif_df["ID_patient"] = modif_df["ID_patient"].astype(str)

# Merge the two DataFrames on subject ID
merged_df = final_folds_df.merge(modif_df, left_on="ID", right_on="ID_patient", how="inner")

# Compare Volume fields
merged_df["Volume_difference"] = merged_df["Volume"] - merged_df["Volume_RF"]

# Find mismatches
discrepant_volumes = merged_df[merged_df["Volume_difference"].abs() > 1e-3]

# Print result
if discrepant_volumes.empty:
    print("✅ All subjects have matching Volume and Volume_RF values.")
else:
    print("❌ Mismatches found in volume values for the following subjects:")
    print(discrepant_volumes[["ID", "Volume", "Volume_RF", "Volume_difference"]])
```

```python
import pandas as pd

# Load the CSV and Excel
final_folds_df = pd.read_csv("final_folds_df.csv")
young_df = pd.read_excel("Young_dataset_RF.xlsx")

# Ensure ID columns are strings
final_folds_df["ID"] = final_folds_df["ID"].astype(str)
young_df["ID_subject"] = young_df["ID_subject"].astype(str)

# Merge the two DataFrames on subject ID
merged_df = final_folds_df.merge(young_df, left_on="ID", right_on="ID_subject", how="inner", suffixes=('_final', '_young'))

# Compare Volume fields
merged_df["Volume_difference"] = merged_df["Volume_final"] - merged_df["Volume_young"]

# Find mismatches
discrepant_volumes = merged_df[merged_df["Volume_difference"].abs() > 1e-3]

# Print result
if discrepant_volumes.empty:
    print("✅ All subjects have matching Volume values between final_folds_df and Young_dataset_RF.")
else:
    print("❌ Mismatches found in volume values for the following subjects:")
    print(discrepant_volumes[["ID", "Volume_final", "Volume_young", "Volume_difference"]])
```
