```python
import random

# Read input file
input_filename = "young_adult_cases.txt"
with open(input_filename, "r") as file:
    lines = file.read().splitlines()

# Extract unique subject IDs (sujeti)
subjects = set(line.split("_RF_")[0] for line in lines)

# Select 2 random subjects for testing
test_subjects = random.sample(subjects, 2)

# Separate train and test cases
train_cases, test_cases = [], []
for line in lines:
    subject = line.split("_RF_")[0]
    if subject in test_subjects:
        test_cases.append(line)
    else:
        train_cases.append(line)

# Write to files
with open("train_cases.txt", "w") as train_file:
    train_file.write("\n".join(train_cases) + "\n")

with open("test_cases.txt", "w") as test_file:
    test_file.write("\n".join(test_cases) + "\n")

print(f"Selected test subjects: {test_subjects}")
print("Files train_cases.txt and test_cases.txt have been created.")
```
