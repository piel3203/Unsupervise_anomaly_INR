```python
import random

```

```python
# Function to read names from a file
def read_names(filename):
    with open(filename, 'r') as file:
        names = file.read().splitlines()
    return names

# Function to split names into training and testing sets
def split_data(names, test_ratio=0.2):
    random.shuffle(names)
    test_size = int(len(names) * test_ratio)
    test_names = names[:test_size]
    train_names = names[test_size:]
    return train_names, test_names

# Function to write names to a file
def write_names(filename, names):
    with open(filename, 'w') as file:
        for name in names:
            file.write(f"{name}\n")
```
